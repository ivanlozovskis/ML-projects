Иногда студенты берут курсы, которые оказываются для них неподъемными. Это может быть как по причинам недостаточной изначальной подготовки, так и по причинам, связанным с низкой мотивацией студента. Для улучшения качества контента полезно выявить причину. На основании имеющихся данных придумайте метрику успеваемости студента. **Обоснуйте ее.** Выявите таких студентов на каждом курсе, чтобы предоставить продюсерам список проблемных студентов. 

```python
#посчитаем процент проваленных дз для каждого студлента,
#затем посчитаем студентов, у которых есть проблемные дз(долго проходят)
arr = []
df_1 = df[df['is_homework']==True]
df_1['dif'] = (df['finish_date'] - df['start_date']).apply(lambda x: x.days)
for i in range(15):
    data = df_1[df_1['title'] == ks.index[i]]
    df_d = data[data['status']=="done"]
    df_w = data[data['status']=='fail']
    d = df_d.groupby(['title','student_id']).count()['course_id']
    w = df_w.groupby(['title','student_id']).count()['course_id']
    c = d.reset_index()
    c['fail'] = w.reset_index()['course_id']
    c['percent'] = 100 * c['fail'] / (c['fail']+c['course_id'])
    c = c[c['percent']>35]
    arr.append(c)
arr
#студенты, у которых проваленных дз больше чем выполненных
```
